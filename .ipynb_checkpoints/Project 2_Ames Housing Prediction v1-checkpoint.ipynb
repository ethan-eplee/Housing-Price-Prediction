```python
#for string columns: print counts of unique values
#for numerical columns, print percentage of zeros 

def summary(df):
    for col in df.columns:
        print('-------- %s--------' % col)
        if df[col].dtype == object:
            print(df[col].value_counts())
        elif len(df[df[col]==0]) >= 1026: #if more than 50% are null values
            print (col, '\n', 'HIGH % zeros:', round((len(df[df[col]==0])/2051)*100),\
                    '\n', 'No. of zeros:', len(df[df[col]==0]))
        else: 
            print(col, '\n', '% zeros:', round((len(df[df[col]==0])/2051)*100))
        print('\n')                
```

#### Correlation of Sale Price with other qualitative features

I have identified the column features that would potentially provide qualitative assessments to each housing, since they have ratings ranking from excellent to poor. The list of column features is as such, in no order of merits: 

- overall_qual (Rates the overall material and finish of the house)
- overall_cond (Rates the overall condition of the house)
- exter_qual (Evaluates the quality of the material on the exterior)
- exter_cond (Evaluates the present condition of the material on the exterior)
- bsmt_qual (Evaluates the height of the basement)
- bsmt_cond (Evaluates the general condition of the basement)
- heating_qc (Heating quality and condition)
- kitchen_qual (Kitchen quality)
- functional (Home functionality)
- fireplace_qu (Fireplace quality)
- garage_qual (Garage quality)
- garage_cond (Garage condition)
- electrical (Electrical system - rated based on type of system)


We need to quantify the features on a scale so that we can check out the correlation of these column features against Sale Price to gain some insights (if any). 

Many of the nominal features had a breakdown from excellent to poor. For these, I changed the variable to ordinal with a score of 1 - 5, 1 for Poor and 5 fo Excellent. 

```python
#converting nominal to ordinal for features with Poor to Excellent rating

ordinal_cols = ['fireplace_qu', 'garage_qual', 'garage_cond', 'bsmt_qual', 'bsmt_cond',
               'heating_qc', 'exter_qual', 'exter_cond', 'kitchen_qual']
for col in ordinal_cols:
    df[col] = df[col].replace({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1}).fillna(0)
    
#column - Electrical (conversion from categorical to ordinal)

df['electrical'] = df['electrical'].map({'SBrkr':5, 'Mix': 4, 'FuseA': 3, 'FuseF':2, 'FuseP':1, np.nan: 0})
df['electrical'] = df['electrical'].astype(int)

#column - functional (conversion from categorical to ordinal)
df['functional'] = df['functional'].map({'Typ':8, 'Min1':7, 'Min2':6, 'Mod':5, 
                                        'Maj1':4, 'Maj2':3, 'Sev':2, 'Sal':1, np.nan: 8}) 
df['functional'] = df['functional'].astype(int)
```

### Model Preparation 

- Create our features matrix (X) and target vector (y)
- Train/test split
- Scale data
- OneHotEncoding on categorical features

```python
train_drop = ['saleprice', 'id', 'year_built', 'yr_sold', 'year_remod/add']
X = df_train.drop(columns = train_drop )
y = df_train['saleprice']

#y = np.log1p(df_train['saleprice'])
```

```python
X.head()
```

```python
X.isnull().sum()
```

```python
X['garage_yr_blt'] = X['garage_yr_blt'].fillna(0)
X['total_bsmt_sf'] = X['total_bsmt_sf'].fillna(0)
X['garage_cars'] = X['garage_cars'].fillna(0)
X['garage_area'] = X['garage_area'].fillna(0)
```

```python
#train test split
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1, random_state = 42)
```

```python
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
```

```python
X_train.info()
```

```python
#dealing with multi-collinearity 
def VIF_calculator(data):
    vif_data = pd.DataFrame()
    num_idx = data.select_dtypes(include = ['int64', 'float64']).columns
    vif_data['feature'] = num_idx
    vif_data['VIF'] = [variance_inflation_factor(data[num_idx].values, i) for i in range(len(num_idx))]
    return vif_data[vif_data['VIF']>10]
```

```python
VIF_calculator(X_train)
```

```python
#Further feature engineering for numerical data
to_drop = ['exter_qual', 'garage_cars', 'garage_yr_blt','totrms_abvgrd', 'bsmt_qual', 'kitchen_qual']
Z_train = X_train.drop(columns = to_drop)
Z_test = X_test.drop(columns = to_drop)
```

```python
#a reduction in VIF values 
VIF_calculator(Z_train)
```

```python
Z_train_column = Z_train.columns
numerical_idx1 = Z_train.select_dtypes(include = ['int64', 'float64']).columns
categorical_idx1 = Z_train.select_dtypes(include = ['object']).columns

ss = StandardScaler()   
Z_train[numerical_idx1] = ss.fit_transform(Z_train[numerical_idx1])
Z_test[numerical_idx1] = ss.transform(Z_test[numerical_idx1])
```

```python
VIF_calculator(Z_train)
#goes to show no features with multicollinearity >10
```

```python
#Encode categorical values
Z_dummy = pd.concat([Z_train, Z_test], axis =0)
OHE = OneHotEncoder(sparse = False, handle_unknown = 'ignore')
trans = ColumnTransformer(transformers = [("Onehot", OHE, categorical_idx1)], remainder= 'passthrough')

Z_dummy = trans.fit_transform(Z_dummy)
```

```python
new_idx = trans.get_feature_names()
Z_dummy = pd.DataFrame(Z_dummy, columns=new_idx)
```

```python
Z_dummy.head()
```

```python
Z_dummy.shape
```

```python
Z_train = Z_dummy[:1493]
Z_test = Z_dummy[1493:1659]
```

```python
print(f'After feature engineering, X training data shape:{Z_train.shape},X testing data shape:{Z_test.shape} while y training data shape:{y_train.shape}, y testing data shape:{y_test.shape}')
```

### Modeling 
- Establish basline score
- Fit linear regression
- Fit lasso/ridge/elastic net with default parameters
- Adjust features so that models can be improved
- Tune hyperparameters
- Identify a production model
- Refine and interpret your production model

```python
#Baseline module

lr = LinearRegression()
lr.fit(Z_train, y_train)
y_train_pred_lr = lr.predict(Z_train)
y_test_pred_lr = lr.predict(Z_test)
```

```python
mse_train = mean_squared_error(y_train, y_train_pred_lr)
mse_test = mean_squared_error(y_test, y_test_pred_lr)

pct_diff = (mse_train - mse_test)/mse_train*100
print(f'mse_train: {mse_train}, mse_test: {mse_test}, {pct_diff}%')
```

```python
# Initialze the estimators
model1 = LinearRegression()
model2 = Ridge()
model3 = Lasso()
model4 = ElasticNet()

# Initialize the hyperparameters for each dictionary
param1 = {}
param1['classifier'] = [model1]

param2 = {}
param2['classifier__alpha'] = [0.01, 0.1, 1, 10, 100]
param2['classifier'] = [model2]

param3 = {}
param3['classifier__alpha'] = [0.01, 0.1, 1, 10, 100]
param3['classifier'] = [model3]

param4 = {}
param4['classifier__alpha'] = [0.01, 0.1, 1, 10, 100]
param4['classifier__l1_ratio'] = np.linspace(0,1,10)
param4['classifier'] = [model4]
```

```python
#create pipeline
pipeline = Pipeline([('classifier', model1)]) 
params = [param1, param2, param3, param4]
```

```python
# run GridsearchCV
gs = GridSearchCV(pipeline, params, cv=10, n_jobs=-1, scoring='neg_mean_squared_error')
gs.fit(Z_train, y_train)
```

```python
gs.best_params_
```

```python
print(abs(gs.best_score_))
```
