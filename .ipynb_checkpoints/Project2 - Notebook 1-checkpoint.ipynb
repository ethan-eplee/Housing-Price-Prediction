There are 2051 rows of data and 81 columns describing features representing the characteristics of the houses sold between 2006 and 2010.

Features can be represented broadly into 5 categories:
- House features (e.g. foundation, masonry type, utilties)
- Size (e.g. no of bedrooms, area of pool, basement square feet)
- Location (e.g. mszoning, neighbourhood)
- Environment (e.g. alley, landslope, land contour)
- Time-related (e.g. year house was built, year garage built, year sold)

The same features can also be classified into three kinds of data: numerical, ordinal and nominal. 
- Numerical data can be measured (e.g. size of house, living area) 
- Ordinal data has predetermined or natural order (e.g. overall quality)
- Nominal data has no natural order or rank (e.g. street type, neighbourhood, roof style)

```python
#explore the data first

def explore_data(df):
    for col in df.columns:
        print(f'------{col}------')
        #for str columns, print value_counts
        if df[col].dtype == np.object:
            print (df[col].value_counts())
            pct_single_value = round(df[col].value_counts()[0]/len(df)*100,2)
            if pct_single_value > 70:
                print('This variable has high percentage of a single value.')
            print('% single value: ', pct_single_value)
            
        #for numerical columns, print % of zeros
        else:
            pct_zeros = round(len(df[df[col]==0])/len(df)*100,2)
            if pct_zeros > 50:
                print('This variable has high percentage of zeros.')
            print('% zeros: ', pct_zeros)
        
        print('\n')
```

## Checking distributions of features using graphs

```python
#plot histograms for categorical columns
def subplot_hist(df, col_list):
    
    # number of rows determined by length of list, 4 columns in plot
    ncols = 4
    nrows =  int(np.ceil(len(col_list)/ncols))
    
    #create subplot
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, nrows*3.5))
    ax = ax.ravel()
    
    #histogram settings
    for i, col in enumerate(col_list):
        ax[i].hist(df[col],bins=15)                           #create hist
        ax[i].set_title(f'{col} distribution', fontsize=14)   #set title
        ax[i].tick_params(labelsize=12)                       #set label size
        ax[i].set_xlabel(col, fontsize=14)                    #set x-label name
    plt.tight_layout()
```

```python
#plot distributions for numerical columns
def subplot_dist(df, col_list):
    
    # number of rows determined by length of list, 4 columns in plot
    ncols = 4
    nrows =  int(np.ceil(len(col_list)/ncols))
    
    #create subplot
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, nrows*3.5))
    ax = ax.ravel()
    
    #dist plot settings
    for i, col in enumerate(col_list):
        sns.distplot(df[col], ax=ax[i])             #create dist
        ax[i].set_title(f'{col} distribution', fontsize=14)   #set title
        ax[i].tick_params(labelsize=12)                       #set label size
        ax[i].set_xlabel(col, fontsize=14)                    #set x-label name
    
    plt.tight_layout()
```

```python
#plot boxplots to see pattern between sale price and categorical feature
def subplot_boxplot(df, col_list):
    
    # number of rows determined by length of list, 4 columns in plot
    ncols = 4
    nrows =  int(np.ceil(len(col_list)/ncols))
    
    #create subplot
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, nrows*3.5))
    ax = ax.ravel()
    
    #dist plot settings
    for i, col in enumerate(col_list):
        sns.boxplot(y=df.saleprice, x=df[col], ax=ax[i], width=0.3,\
                    color='skyblue')                                   #create dist
        ax[i].set_title(f'{col} vs saleprice', fontsize=14)            #set title
        ax[i].tick_params(labelsize=12)                                #set label size
        ax[i].set_xlabel(col, fontsize=14)                             #set x-label name
        ax[i].set_ylabel('saleprice', fontsize=14)                     #set y-label name
    plt.tight_layout()
```

```python
#plot scatterplots to see correlations between sale price and numerical features
def subplot_scatter(df, col_list):
    
    # number of rows determined by length of list, 4 columns in plot
    ncols = 4
    nrows =  int(np.ceil(len(col_list)/ncols))
    
    #create subplot
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, nrows*3.5))
    ax = ax.ravel()
    
    #dist plot settings
    for i, col in enumerate(col_list):
        sns.regplot(y=df.saleprice, x=df[col], ax=ax[i],\
                    scatter_kws={'facecolors':'skyblue','edgecolor':'skyblue'},\
                    line_kws = {'color':'grey'})                       #create dist
        ax[i].set_title(f'{col} vs saleprice', fontsize=14)            #set title
        ax[i].tick_params(labelsize=12)                                #set label size
        ax[i].set_xlabel(col, fontsize=14)                             #set x-label name
        ax[i].set_ylabel('saleprice', fontsize=14)                     #set y-label name
    plt.tight_layout()
```
